# Problema fácil - 500 nós e 64624 arestas

Para entendermos o comportamento inicial do algoritmo geramos uma visualização que varia entre as opções de números de formigas (*ants*) e gerações (*gen*), mantendo a taxa de evaporação fixa em *0.05*. Esse valor de evaporação foi escolhido pois sabemos pela literatura do ACO que quando a atualização dos feromônios do melhor caminho é feita de forma elitista o algoritmo tende a gerar melhores resultados com taxas baixas de evaporação, caso contrário o otimizador favorece muito *Explotation* frente à *Exploration*, otimizando apenas no ótimo local e não buscando pelo ótimo global

```python
easy_50_50_005 = pd.read_csv('analysis/easy-50-50-0.05',nrows=50)
easy_50_100_005= pd.read_csv('analysis/easy-50-100-0.05', nrows=100)
easy_50_500_005 = pd.read_csv('analysis/easy-50-500-0.05', nrows=500)
easy_100_50_005 = pd.read_csv('analysis/easy-100-50-0.05', nrows=50)
easy_100_100_005 = pd.read_csv('analysis/easy-100-100-0.05', nrows=100)
easy_100_500_005 = pd.read_csv('analysis/easy-100-500-0.05', nrows=500)
easy_200_50_005 = pd.read_csv('analysis/easy-200-50-0.05', nrows=50)
easy_200_100_005 = pd.read_csv('analysis/easy-200-100-0.05', nrows=100)
easy_200_500_005 = pd.read_csv('analysis/easy-200-500-0.05', nrows=500)
easy_500_50_005 = pd.read_csv('analysis/easy-500-50-0.05', nrows=50)
easy_500_100_005 = pd.read_csv('analysis/easy-500-100-0.05', nrows=100)
easy_500_500_005 = pd.read_csv('analysis/easy-500-500-0.05', nrows=500)
```

```python
easy_200_200_005 = pd.read_csv('analysis/easy-200-200-0.05',nrows=200)
easy_200_200_010 = pd.read_csv('analysis/easy-200-200-0.10', nrows=200)
easy_200_200_020 = pd.read_csv('analysis/easy-200-200-0.20', nrows=200)
easy_200_200_030 = pd.read_csv('analysis/easy-200-200-0.30', nrows=200)
easy_200_200_050 = pd.read_csv('analysis/easy-200-200-0.50', nrows=200)
easy_max_by_evap = pd.DataFrame({
   "Evaporation Rate: 0.05" : easy_200_200_005["best"],
   "Evaporation Rate: 0.10" : easy_200_200_010["best"],
   "Evaporation Rate: 0.20" : easy_200_200_020["best"],
   "Evaporation Rate: 0.30" : easy_200_200_030["best"],
   "Evaporation Rate: 0.50" : easy_200_200_050["best"],
  })

sns.set(rc={'figure.figsize':(15, 10)})
sns.lineplot(data=easy_max_by_evap[[
   "Evaporation Rate: 0.05",
   "Evaporation Rate: 0.10",
   "Evaporation Rate: 0.20",
   "Evaporation Rate: 0.30",
   "Evaporation Rate: 0.50",
    ]])
plt.xlabel("Gerações")
plt.ylabel("Max Clique")
plt.title("Max Clique dos melhores indivíduos por taxa de evaporação")
plt.xlim([0, 200])
# plt.ylim([6.0, 7.0])
plt.show()
```

# Conclusão 

Para todos os três problemas de avaliação observamos que quanto maior o valor dos parâmetros *ants* e *gen* melhor são os resultados, porém a melhoria dos resultados caí quando esses valores passam de 200. A execução de experimentos com mais de 500 formigas ou gerações é inviável visto que nos problemas difíceis para esses valores cada execução demorou cerca de cinco horas. Apesar disso, é seguro dizer que o aumento além do valor de 500 desses parâmetros geraria aumentos cada vez menores nos resultados até não haver mais melhoria.

No que diz respeito à taxa de evaporação, como discutido anteriormente, observamos o que era esperado nos problemas difíceis. Para um ACO que atualiza os melhores caminhos de forma elitista é benéfico usar uma taxa de evaporação mais baixa para favorecer a exploração mais abrangente do espaço de soluções. Já para o problema fácil observamos taxas ligeiramente maiores de evaporação gerando resultados superiores. Isso se deve à capacidade do algoritmo implementado de encontrar o máximo global, provavelmente devido ao número reduzido de nós no gráfico, visto que com volume de formigas superior ao de nós a exploração é favorecida. Sendo assim para esse problema o aumento da *exploitation* por meio do aumento da taxa de evaporação favorece convergência mais rápida para o ótimo global.
