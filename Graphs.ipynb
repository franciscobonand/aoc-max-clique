```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
easy_50_50_005 = pd.read_csv('analysis/easy-50-50-005.csv',nrows=50)
easy_50_100_005= pd.read_csv('analysis/easy-50-100-0.05', nrows=100)
easy_50_500_005 = pd.read_csv('analysis/easy-50-500-0.05', nrows=500)
easy_100_50_005 = pd.read_csv('analysis/easy-100-50-0.05', nrows=50)
easy_100_100_005 = pd.read_csv('analysis/easy-100-100-0.05', nrows=100)
easy_100_500_005 = pd.read_csv('analysis/easy-100-500-0.05', nrows=500)
easy_500_50_005 = pd.read_csv('analysis/easy-500-50-0.05', nrows=50)
easy_500_100_005 = pd.read_csv('analysis/easy-500-100-0.05', nrows=100)
easy_500_500_005 = pd.read_csv('analysis/easy-500-500-0.05', nrows=500)
```

```python
easy_50_50_005.head()
```

```python
min_tour_by_gen = pd.DataFrame({
    "Ants: 50 & Gen: 50": easy_50_50_005["best"],
    "Ants: 50 & Gen: 100": easy_50_100_005["best"],
    "Ants: 50 & Gen: 500": easy_50_500_005["best"],
    "Ants: 100 & Gen: 50": easy_100_500_005["best"],
    "Ants: 100 & Gen: 100": easy_100_100_005["best"],
    "Ants: 100 & Gen: 500": easy_100_500_005["best"],
    "Ants: 500 & Gen: 50": easy_500_50_005["best"],
    "Ants: 500 & Gen: 100": easy_500_100_005["best"],
    "Ants: 500 & Gen: 500": easy_500_500_005["best"],
  })

sns.set(rc={'figure.figsize':(15, 10)})
sns.lineplot(data=min_tour_by_gen[[
    "Ants: 50 & Gen: 50",
    "Ants: 50 & Gen: 100",
    "Ants: 50 & Gen: 500",
    "Ants: 100 & Gen: 50",
    "Ants: 100 & Gen: 100",
    "Ants: 100 & Gen: 500",
    "Ants: 500 & Gen: 50",
    "Ants: 500 & Gen: 100",
    "Ants: 500 & Gen: 500",
    ]])
plt.xlabel("Geraes")
plt.ylabel("Max Clique")
plt.title("Max Clique dos melhores indivduos por gerao")
plt.xlim([0, 500])
# plt.ylim([6.0, 7.0])
plt.show()
```
